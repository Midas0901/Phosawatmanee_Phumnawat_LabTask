```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

print("✅ Basic libraries imported!")
```

```python
# LOAD THE DATA
df = pd.read_csv("/content/sample_data/music_streams.csv")

print("✅ Data loaded successfully!")
print("Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())
```

```python
#ALTAIR IMPORT & SETUP
import altair as alt

alt.data_transformers.enable('default', max_rows=None)

print("✅ Altair ready for interactive charts!")
```

```python
# 1. Bar Chart - Average Energy by Genre
bar_chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('genre:N', sort='-y', title='Genre'),
    y=alt.Y('mean(energy):Q', title='Average Energy'),
    color=alt.Color('mean(energy):Q', scale=alt.Scale(scheme='viridis')),
    tooltip=['genre:N', alt.Tooltip('mean(energy):Q', format='.3f'), alt.Tooltip('count()')]
).properties(
    title='Average Energy by Music Genre',
    width=650,
    height=400
).interactive()

bar_chart
```

```python
# 2. Scatter Plot - Tempo vs Energy
scatter = alt.Chart(df).mark_circle(size=70, opacity=0.8).encode(
    x=alt.X('tempo_bpm:Q', title='Tempo (BPM)'),
    y=alt.Y('energy:Q', title='Energy'),
    color='genre:N',
    tooltip=['title', 'artist', 'genre', 'tempo_bpm', 'energy']
).properties(
    title='Tempo vs Energy by Genre',
    width=700,
    height=450
).interactive()

scatter
```
