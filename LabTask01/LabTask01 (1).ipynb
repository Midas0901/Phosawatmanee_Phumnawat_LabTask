### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

2. Subtract the average age from every age value.

3. Subtract a custom array from two selected columns in the dataset.

```python
import numpy as np

# Create a new column that converts fares from USD to Euros (Assuming conversion rate 1 USD = 0.92 EUR)
df['Fare_EUR'] = df['Fare'] * 0.92

# Subtract the average age from every age value
mean_age = df['Age'].mean()
df['Age_Centered'] = df['Age'] - mean_age

# Subtract a custom array from two selected columns in the dataset
# Let's use 'SibSp' (Siblings/Spouses) and 'Parch' (Parents/Children) as our two columns
# We create a custom numpy array of penalties/offsets to subtract from each row
custom_array = np.array([1, 1])
df[['SibSp_Modified', 'Parch_Modified']] = df[['SibSp', 'Parch']] - custom_array

df[['Fare', 'Fare_EUR', 'Age', 'Age_Centered', 'SibSp', 'SibSp_Modified']].head()
```

### Part 6: Applying Functions ###

1. Create a new column with the number of characters in each name.

2. Calculate the difference between fare and age for each row.

3. Round all numerical values in the dataset to one decimal place.



```python
# Create a new column with the number of characters in each name
df['Name_Length'] = df['Name'].apply(len)

# Calculate the difference between fare and age for each row
df['Fare_Age_Diff'] = df['Fare'] - df['Age']

# Round all numerical values in the dataset to one decimal place
# We select numeric types first to avoid errors on text columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].round(1)

df[['Name', 'Name_Length', 'Fare', 'Age', 'Fare_Age_Diff']].head()
```

### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
# Convert categorical values in the gender column to numeric codes (male: 0, female: 1)
# Note: The Titanic dataset column name for gender is typically 'Sex'
df['Sex_Code'] = df['Sex'].map({'male': 0, 'female': 1})

# Create a new column that categorizes age into groups (child, adult, senior)
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(categorize_age)

df[['Sex', 'Sex_Code', 'Age', 'Age_Group']].head(10)
```

```python
import csv
import urllib.request
import time
import pandas as pd
import numpy as np

print("--- Step 1: Downloading and preparing data ---")
# 1. Load data for Pure Python (List of Dictionaries)
url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"
response = urllib.request.urlopen(url)
lines = [line.decode('utf-8') for line in response.readlines()]
reader = csv.DictReader(lines)
titanic_list = list(reader)

# 2. Load data for Pandas (DataFrame)
df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv")
print("Data successfully loaded for both versions.\n")


print("--- Step 2: Running Pure Python Version ---")
start_pure = time.perf_counter()

# Task 2: Selection (Select Name, Age, Survived)
selected_data_pure = []
for row in titanic_list:
    selected_data_pure.append({
        'Name': row['Name'],
        'Age': row['Age'],
        'Survived': row['Survived']
    })

# Task 3: Filtering (Age > 60)
filtered_data_pure = []
for row in titanic_list:
    if row['Age'] and float(row['Age']) > 60:
        filtered_data_pure.append(row)

# Task 5: Arithmetic (Convert Fare USD to EUR)
for row in titanic_list:
    if row['Fare']:
        row['Fare_EUR'] = float(row['Fare']) * 0.92
    else:
        row['Fare_EUR'] = None

end_pure = time.perf_counter()
pure_duration = end_pure - start_pure
print(f"Pure Python finished in {pure_duration:.6f} seconds.")


print("--- Step 3: Running Pandas Version ---")
start_pandas = time.perf_counter()

# Task 2: Selection
pandas_selection = df[['Name', 'Age', 'Survived']]

# Task 3: Filtering
pandas_filtering = df[df['Age'] > 60]

# Task 5: Arithmetic
df['Fare_EUR'] = df['Fare'] * 0.92

end_pandas = time.perf_counter()
pandas_duration = end_pandas - start_pandas
print(f"Pandas finished in {pandas_duration:.6f} seconds.\n")


print("=== Execution Time Comparison ===")
print(f"Pure Python: {pure_duration:.6f} seconds")
print(f"Pandas:      {pandas_duration:.6f} seconds")
print("-" * 34)

if pure_duration < pandas_duration:
    speedup = pandas_duration / pure_duration
    print(f"Pure Python was faster by {speedup:.2f}x for this small dataset.")
else:
    speedup = pure_duration / pandas_duration
    print(f"Pandas was faster by {speedup:.2f}x.")
```
